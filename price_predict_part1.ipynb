## Data Collection

```python
# Import library
import pandas as pd
import numpy as np
```

```python
# Read the online file by the URL, and assign it to variable "df"
other_path = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DA0101EN-SkillsNetwork/labs/Data%20files/auto.csv"
df = pd.read_csv(other_path, header=None)
```

```python
df.head()
```

To better describe our data, we add a header

```python
#Create header list
headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
```

```python
#Replace header and check our dataframe:
df.columns = headers
df.head()
```

```python
# Replace the '?' symbol with NaN so the dropna() can remove the missing values:
df1 = df.replace('?', np.NaN)
```

```python
#Since we are predicting the price, drop the missing values along the column "price"
df = df1.dropna(subset = ['price'], axis = 0)
df.head()
```

### Dealing with missing data
To deal with missing data, we can either drop or replace them with mean, frequency, or other functions.

```python
#Calculate the mean value for the "normalized-losses" column
avg_norm_loss = df['normalized-losses'].astype('float').mean(axis = 0)
print("Average of normalized-losses:", avg_norm_loss)
```

```python
#Replace "NaN" with mean value in "normalized-losses" column
df['normalized-losses'].replace(np.nan, avg_norm_loss, inplace= True )
```

```python
#Calculate the mean value for the "bore" column
avg_bore = df['bore'].astype('float').mean(axis = 0)
print("Average of bore:", avg_bore)
```

```python
#Replace "NaN" with the mean value in the "bore" column
df["bore"].replace(np.nan, avg_bore, inplace = True)
```

```python
#Calculate the mean value of "stroke" column
avg_stroke = df['stroke'].astype('float').mean(axis = 0)
print("Average of stroke: ",avg_stroke)
```

```python
#Replace "NaN" with the mean value in the "stroke" column
df['stroke'].replace(np.nan, avg_stroke, inplace = True)
```

```python
#We will repeat the same process with "horsepower" and "peak-rpm"
avg_horsepower = df['horsepower'].astype('float').mean(axis = 0)
print("Average horsepower: ", avg_horsepower)
```

```python
df['horsepower'].replace(np.nan, avg_horsepower, inplace = True)
```

```python
avg_peakrpm = df['peak-rpm'].astype('float').mean(axis = 0)
print("Average peak rpm: ", avg_peakrpm)
```

```python
df['peak-rpm'].replace(np.nan, avg_peakrpm, inplace = True)
```

```python
df['num-of-doors'].value_counts()
```

```python
#As four door is the most common type, we replace the missing 'num-of-doors' values by the most frequent
df['num-of-doors'].replace(np.nan, "four", inplace = True)
```

```python
df.head()
```

Now that the data set has no missing values

```python
df.dtypes
```

```python
#Convert data types to proper format
df[["bore", "stroke"]] = df[["bore", "stroke"]].astype("float")
df[["normalized-losses"]] = df[["normalized-losses"]].astype('int')
df[['price']] = df[['price']].astype("float")
df[["peak-rpm"]] = df[['peak-rpm']].astype("float")
```

```python
df.dtypes
```

Now, the data is clean with no missing value and with all data in its proper format

### Data standardization

Standardization is the process of transforming data into a common format, allowing the researcher to make the meaningful comparison.

In this case, the fuel consumption columns "city-mpg" and "highway-mpg" are represented by mpg (miles per gallon) unit, we will transform mpg into L/100km (L/100km = 235/mpg)

```python
#Convert mpg to L/100km by mathematical operation (235 divided by mpg)
df['city-L/100km'] = 235/df['city-mpg']
df.head()
```

```python
df['highway-L/100km'] = 235/df['highway-mpg']
```

```python
df.head()
```

### Data Normalization

Normalization the process of transforming values of several variables into a similar range. 

We will scale the columns "length", "width", and "height" to normalize those variables so their value ranges from 0 to 1.

```python
#replace (original value) by (original value)/(maximum value)
df['length'] = df['length']/df['length'].max()
df['width'] = df['width']/df['width'].max()
df['height'] = df['height']/df['height'].max()
```

## Binning
In the data set, 'horsepower' is a real valued variable ranging from 48 to 288 and it has 59 unique values.

We will use the cut method to segment the horsepower column into 3 bins: high horsepower, medium horsepower, and little horse power

```python
df['horsepower'].nunique()
```

```python
df['horsepower'] = df['horsepower'].astype(int, copy = True)
```

```python
#Plot the histogram of horsepower to see the distribution od horsepower
%matplotlib inline
import matplotlib as plt
from matplotlib import pyplot
plt.pyplot.hist(df['horsepower'])

#Set x/y labels and plot title
plt.pyplot.xlabel("horserpower")
plt.pyplot.ylabel("count")
plt.pyplot.title("horsepower bins")
```

```python
#Find 3 bins of equal size bandwidth
bins = np.linspace(min(df["horsepower"]), max(df["horsepower"]), 4)
bins
```

```python
group_names = ['Low', 'Medium', 'High']
```

```python
df['horsepower-binned'] = pd.cut(df['horsepower'], bins, labels= group_names, include_lowest= True)
df[['horsepower', 'horsepower-binned']].head(20)
```

```python
df['horsepower-binned'].value_counts()
```

```python
#Plot the distribution of each bin
%matplotlib inline
import matplotlib as plt
from matplotlib import pyplot
pyplot.bar(group_names, df["horsepower-binned"].value_counts())

# set x/y labels and plot title
plt.pyplot.xlabel("horsepower")
plt.pyplot.ylabel("count")
plt.pyplot.title("horsepower bins")
```

### Dummy Variable

The column "fuel-type" has two unique values: "gas" or "diesel". We use the Panda method 'get_dummies' to assign numerical values to different categories of fuel type

```python
# Get the indicator variables and assign it to data frame "dummy_variable_1"
dummy_variable_1 = pd.get_dummies(df["fuel-type"])
dummy_variable_1
```

```python
#Change the column names for clarity:
dummy_variable_1.rename(columns = {'gas': 'fuel-type-gas', 'diesel': 'fuel-type-diesel'}, inplace = True)
dummy_variable_1.head()
```

```python
# Merge data frame "df" and "dummy_variable_1"
df = pd.concat([df, dummy_variable_1], axis = 1)

# Drop orginnal column "fuel-type" from "df"
df.drop("fuel-type", axis = 1, inplace = True)
```

```python
df.head()
```

```python
#Create an indicator variable for the column "aspiration"
dummy_variable_2 = pd.get_dummies(df['aspiration'])
dummy_variable_2.head()
```

```python
dummy_variable_2.rename(columns={'std': 'aspiration-std', 'turbo': 'aspiration-turbo'}, inplace= True)
dummy_variable_2.head()
```
